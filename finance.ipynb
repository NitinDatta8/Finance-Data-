```python
# First we will see Gender of our respondenets
sns.countplot(df['gender'])
```

```python
# Now we will see Age of our respondents
sns.countplot(df['age'])


'''We can see from the below graph that most of our respondents were in the ages 24-32 and we will look further
their habits in the finanace field '''
```

```python
sns.countplot(x=df['age'],hue=df['gender'],palette='spring')

# From the below plot we can see that most of the males were between 24-32
```

```python
# Now lets see age wise  whether people invest in INVESTMENT AVENUES and STOCK MARKET 
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(x=df['age'],hue=df['Investment_Avenues'],palette='summer')
plt.title('INVESTMENT AVENUE')
plt.subplot(1, 2, 2)
sns.countplot(x=df['age'],hue=df['Stock_Marktet'],palette='hot')
plt.title('STOCK MARKET')
plt.show()


# Only 3 people dont invest and 5 people dont go to stock market
```

```python
# Now lets see these same factors of INVESTMENT_AVENUE and STOCK_MARKET with gender

plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(x=df['gender'],hue=df['Investment_Avenues'],palette='summer')
plt.title('INVESTMENT AVENUE')
plt.subplot(1, 2, 2)
sns.countplot(x=df['gender'],hue=df['Stock_Marktet'],palette='hot')
plt.title('STOCK MARKET')
plt.show()


# Mostly females dont invest or go to stocks 
```

```python
# Factors affecting Investing 
sns.countplot(x=df['Factor'],palette='coolwarm')

# Returns is the most influential Factor considered while investing which is followed by risk 
```

## The above trend shows us that people are less worried about risks involved and more worried about the money which they will get back 

```python
# Lets see which Age groups take the highest risk 
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(x=df['age'],hue=df['Factor'],palette='magma')
plt.title('Age')
plt.subplot(1, 2, 2)
sns.countplot(x=df['gender'],hue=df['Factor'],palette='Oranges')
plt.title('Gender')
plt.show()
```

### We can see that 24-32 aged people try to focus more on returns than the risks involved, which might be because they want to earn more and more also high number of males are dependent on Returns
